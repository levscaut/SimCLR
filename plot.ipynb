```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
resnet18 = pd.read_csv('results/resnet18.csv')
resnet50 = pd.read_csv('results/resnet50.csv')
resnet18_supervised = pd.read_csv('results/resnet18_supervised.csv')
resnet50_supervised = pd.read_csv('results/resnet50_supervised.csv')
```

```python
simclr_df = resnet18.join(resnet50, lsuffix='_resnet18', rsuffix='_resnet50')
display(simclr_df)
```

```python
sns.set_theme(style='white')
plt.figure(figsize=(8, 8))
# ç»˜åˆ¶Lossæ›²çº¿
sns.lineplot(x='Step_resnet18', y='Value_resnet18', data=simclr_df, label='ResNet18')
sns.lineplot(x='Step_resnet50', y='Value_resnet50', data=simclr_df, label='ResNet50')

# æ·»åŠ æ ‡é¢˜å’Œæ ‡ç­¾
plt.title('Training Loss During Pre-Training of SimCLR')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
supervised_df = resnet18_supervised.join(resnet50_supervised, lsuffix='_resnet18', rsuffix='_resnet50')
sns.set_theme(style='white')
plt.figure(figsize=(8, 8))
# ç»˜åˆ¶Lossæ›²çº¿
sns.lineplot(x='Step_resnet18', y='Value_resnet18', data=supervised_df, label='ResNet18')
sns.lineplot(x='Step_resnet50', y='Value_resnet50', data=supervised_df, label='ResNet50')

# æ·»åŠ æ ‡é¢˜å’Œæ ‡ç­¾
plt.title('Training Loss During Supervised Training')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
